```python
import numpy as np    # for using nd-arrays and mathematical computations
from sklearn import linear_model    # to use buildin Linear Regression function
import matplotlib.pyplot as plt    # for plotting graphs
```

```python
# Creating an array which contains 2 arrays
#     1. Areas [ a1, a2, a3,... ]
#     2. Prices [ p1, p2, p3,... ]

input = np.array([[2600, 3000, 3200, 3600, 4000], [550000, 565000, 610000, 680000, 725000]])
```

```python
# PLotting the input data 

%matplotlib inline

plt.title('Input Data Plot')
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(input[0],input[1],color='red',marker='+')
```

<h1> Now we have to perform linear regression over this data. </h1>
<h2> <u> For this we use the sklearn library </u> </h2>

 <h3>
    The LinearRegression function takes 2 elements as the input.
    
        1. Arrays of different Features [ [Feature1_1, Feature2_1, Feature3_1,...],
                                       [Feature1_2, Feature2_2, Feature3_2,...],
                                       [Feature1_3, Feature2_3, Feature3_3,...],
                                       ...]
         
        2. (-1, 1) shape 2d-array of Targets
</h3>

```python
areas = input[0].reshape(-1, 1)
prices = input[1].reshape(-1, 1)

print('areas :')
print(areas)

print()

print('prices :')
print(prices)
```

```python
#Running the Linear Regression Model
reg = linear_model.LinearRegression()
reg.fit(areas, prices)
```

```python
# Checking the trained model
reg.predict([[3300]])
```

```python
# taking out our parameters from the model
# y = mx + c   =>  price = m * area + c

m = reg.coef_[0][0]    # reg.coef_  returns a 2d-array so we extract the arr[0][0] element
c = reg.intercept_[0]  # reg.intercept_  returns a 1d array

print("m = ", m)
print("c = ", c)
```

## We got our model
### `predicted_price = m * input_area + c`

```python
## function to draw the linear reggesion line using slope and intercept

def abline(slope, intercept):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--')
    
slope = m
intercept = c
```

```python
# PLotting features and target graph 

%matplotlib inline
plt.title('Feature vs Target')
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(input[0],input[1],color='red',marker='+')

abline(slope, intercept)
```

### `This graph proves that yes it is a best fit line for our linear regression model`
